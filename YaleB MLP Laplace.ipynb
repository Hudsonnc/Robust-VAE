```python
# I use the code snippet from https://stackoverflow.com/questions/7368739/numpy-and-16-bit-pgm/7369986
# It read PGM image buffer string and return a numpy array
def read_pgm2(buffer, byteorder='>'):
    """Return image data from a raw PGM file as numpy array.
    Format specification: http://netpbm.sourceforge.net/doc/pgm.html
    """
    #with open(filename, 'rb') as f:
    #    buffer = f.read()
    try:
        header, width, height, maxval = re.search(
            b"(^P5\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n]\s)*)", buffer).groups()
    except AttributeError:
        raise ValueError("Not a raw PGM file: '%s'" % filename)
    #print 'width',width, 'height', height
    
    return np.frombuffer(buffer,
                            dtype='u1' if int(maxval) < 256 else byteorder+'u2',
                            count=int(width)*int(height),
                            offset=15#len(header)
                            ).reshape((int(height), int(width)))
```

# RVAE

```python
import networks
from RobustVAE import RobustVAE

import numpy as np
import tensorflow as tf

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data_dim = X.shape[1]
latent_dim = 100

random_seed = 1234

encoder = networks.MLP(data_dim, latent_dim, 3, [300,300,300],  linear_out=True, random_seed = random_seed)
decoder = networks.MLP(latent_dim, data_dim, 3, [300,300,300],  linear_out=False, random_seed = random_seed)
rvae = RobustVAE(encoder, decoder)
```

```python
n_sparse = 90
rvae.train(X, n_sparse = n_sparse, iters=20000, batch_size=100, lr=1e-2)
```

```python
zhat = encoder(X)
zhat = rvae.sess.run(zhat)
plt.scatter(zhat[:,-1], zhat[:,-2])
```

```python
plt.imshow(X[0].reshape((shapeX[1],shapeX[2])), plt.cm.gray)
```

```python
xhat = decoder(encoder(X))
xhat = rvae.sess.run(xhat)
plt.imshow(xhat[0].reshape((shapeX[1],shapeX[2])), plt.cm.gray)
```

```python
z = encoder(X)
z = rvae.sess.run(z)
z0 = np.copy(z[0])
z0[-n_sparse:] = 0
```

```python
x0new = decoder(np.array([z0]))
x0new = rvae.sess.run(x0new)
plt.imshow(x0new[0].reshape((shapeX[1],shapeX[2])), plt.cm.gray)
```
